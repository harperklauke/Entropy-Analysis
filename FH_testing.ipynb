```python
'T-test'
print("Full\n", stats.ttest_ind(fh_full_df['Reg1AvgEntropy'],mh_full_df['Reg1AvgEntropy'],equal_var=True,
                     alternative = 'less'))
print("Cropped\n", stats.ttest_ind(fh_df['Reg1AvgEntropy'],mh_df['Reg1AvgEntropy'],equal_var=True,
                     alternative = 'less'))
print("Full Gray\n", stats.ttest_ind(fh_full_gray_df['Reg1AvgEntropy'],mh_full_gray_df['Reg1AvgEntropy'],equal_var=True,
                     alternative = 'less'))
print("Cropped Gray\n", stats.ttest_ind(fh_gray_df['Reg1AvgEntropy'],mh_gray_df['Reg1AvgEntropy'],equal_var=True,
                     alternative = 'less'))
```

```python
'Mann Whitney'
print("Full\n", stats.mannwhitneyu(fh_full_df['Reg1AvgEntropy'],mh_full_df['Reg1AvgEntropy'],
                     alternative = 'less'))
print("Cropped\n", stats.mannwhitneyu(fh_df['Reg1AvgEntropy'],mh_df['Reg1AvgEntropy'],
                     alternative = 'less'))
print("Full Gray\n", stats.mannwhitneyu(fh_full_gray_df['Reg1AvgEntropy'],mh_full_gray_df['Reg1AvgEntropy'],
                     alternative = 'less'))
print("Cropped Gray\n", stats.mannwhitneyu(fh_gray_df['Reg1AvgEntropy'],mh_gray_df['Reg1AvgEntropy'],
                     alternative = 'less'))
```

```python
print(stats.f_oneway(fh_df['Reg1AvgEntropy'],mh_df['Reg1AvgEntropy']))
print(stats.f_oneway(fh_gray_df['Reg1AvgEntropy'],mh_gray_df['Reg1AvgEntropy']))
print(stats.f_oneway(fh_full_df['Reg1AvgEntropy'],mh_full_df['Reg1AvgEntropy']))
print(stats.f_oneway(fh_full_gray_df['Reg1AvgEntropy'],mh_full_gray_df['Reg1AvgEntropy']))
```
